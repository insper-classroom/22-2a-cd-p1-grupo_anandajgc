```python
import re 

def cleanup(text):
    punctuation = '(:),'
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    text_subbed = text_subbed.replace('\n',' ')
    text_subbed = text_subbed.replace('"',' ')
    text_subbed = text_subbed.replace("'",' ')
    return text_subbed
```

```python
# LaPlace
def LaPlace0(palavra):
    if palavra not in list_train0:
        x = 1 / (len(list_train0) + len(list_train_norep))
    else:
        x = fa_train0[palavra] / fa_train0.sum()
    return x

def LaPlace1(palavra):
    if palavra not in list_train1:
        x = 1 / (len(list_train1) + len(list_train_norep))
    else:
        x = fa_train1[palavra] / fa_train1.sum()
    return x

def LaPlace2(palavra):
    if palavra not in list_train2:
        x = 1 / (len(list_train2) + len(list_train_norep))
    else:
        x = fa_train2[palavra] / fa_train2.sum()
    return x
```

```python
# Naive Bayes
def cprob0(palavra):
    p = 1
    p *= LaPlace0(palavra)
    return p

def cprob1(palavra):
    p = 1
    p *= LaPlace1(palavra)
    return p

def cprob2(palavra):
    p = 1
    p *= LaPlace2(palavra)
    return p

def NaiveBayes(noticia):
    prob_not0 = cprob0(noticia)
    prob_not1 = cprob1(noticia)
    prob_not2 = cprob2(noticia)
    
    prob_0not = prob_not0 * prob0
    prob_1not = prob_not1 * prob1
    prob_2not = prob_not2 * prob2
    
    if prob_1not > prob_0not and prob_1not > prob_2not:
        return 1
    elif prob_2not > prob_1not and prob_2not > prob_0not:
        return 2
    else:
        return 0
```
